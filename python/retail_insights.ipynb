This notebook explores multiple business insights using Python visualizations.


```python
import pandas as pd
import matplotlib.pyplot as plt

sales = pd.read_csv('sales.csv')
sales['sale_date'] = pd.to_datetime(sales['sale_date'])
```

## 1. Revenue by Store (Top vs Bottom)

```python
rev_store = sales.groupby('store_id')['revenue'].sum().sort_values()
plt.figure()
rev_store.tail(10).plot(kind='bar')
plt.title('Top 10 Stores by Revenue')
plt.show()
```

## 2. Monthly Revenue Seasonality

```python
sales['month'] = sales['sale_date'].dt.month
monthly = sales.groupby('month')['revenue'].sum()
plt.figure()
monthly.plot()
plt.title('Monthly Revenue Trend')
plt.show()
```

## 3. Average Order Value by Store

```python
aov = sales.groupby('store_id')['revenue'].mean()
plt.figure()
aov.plot(kind='hist', bins=20)
plt.title('Distribution of Average Order Value')
plt.show()
```

## 4. Units Sold Distribution

```python
plt.figure()
sales['units_sold'].plot(kind='hist', bins=30)
plt.title('Units Sold Distribution')
plt.show()
```

## 5. Store Revenue Contribution

```python
contrib = rev_store / rev_store.sum()
plt.figure()
contrib.plot()
plt.title('Revenue Contribution by Store')
plt.show()
```
